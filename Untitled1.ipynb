```python
import torch
import torchvision
from torch.utils.data.dataset import Dataset
import PIL
from pathlib import PosixPath
from pathlib import Path
```

```python
class HorizonSquare(Dataset):
    def __init__(self, root, transform=None, ):
        self.root = root
        self.transform = transform
        
        self.images_path = Path(self.root).joinpath('images').glob('*.jpg')
        self.images_files = sorted(list(self.images_path))
        
        self.masks_sky_path = Path(self.root).joinpath('masks/sky').glob('*.png')
        self.masks_sky_files = sorted(list(self.masks_sky_path))
        
        self.masks_land_path = Path(self.root).joinpath('masks/land').glob('*.png')
        self.masks_land_files = sorted(list(self.masks_land_path))
                                  
    def __len__(self):
        return len(self.images_files)
    
    def _load_image(self, path: PosixPath):
        return PIL.Image.open(str(path))
    
    def __getitem__(self, idx):
        img_path = self.images_files[idx]
        img_ori = self._load_image(img_path)
        
        sky_path = self.masks_sky_files[idx]
        img_sky = self._load_image(sky_path)
        
        land_path = self.masks_sky_files[idx]
        img_land = self._load_image(land_path)
        
        return img_ori, img_land, img_sky
```
